## Ejercicio 11

Para todos los comentarios de revisión de contenido que tengan más de 20
ocurrencias realice una matriz cuyas columnas sean esos comentarios y de índice
los usuarios/ips con valores: True si ese usuario realizó ese comentario, sino False


```python
import pandas as pd
import numpy as np
```

```python
contents = pd.read_csv('contents.csv')
contents.head()
```

Para realizar el ejercicio, unifico las columnas de revisor_username y revisor ip dado que cuando una es NaN la otra contiene valores.

```python
unified_col = contents['revisor_username'].fillna(contents['revisor_ip'])
unified_col
```

Agrego la columna al final de mi dataframe original y solo conservo aquellas columnas que necesito para trabajar.

```python
contents['username/ips'] = unified_col
```

```python
contents_reduced = contents[['username/ips','revisor_comment']]
contents_reduced.head()
```

```python
# Filtro aquellas filas que pertenecen a aquellos comentarios que poseen mas de 20 apariciones.
comments = contents_reduced['revisor_comment'].value_counts()
comments_needed = comments[comments > 20]
contents_reduced_filtered = contents_reduced.loc[contents_reduced['revisor_comment'].isin(comments_needed.index)]
contents_reduced_filtered.head()
```

```python
contents_reduced_filtered.count()
```

```python
# Como vemos, tenemos un username/ips que posee NaN y dado que es uno solo lo droppeo.
cleaned_contents = contents_reduced_filtered.dropna()
cleaned_contents.count()
```

```python
# Genero un dataframe usando el pivot-table y lleno los valores con 0 y las apariciones de los usuarios.
pivot_contents = cleaned_contents.pivot_table(index='username/ips',columns=['revisor_comment'],aggfunc={'username/ips':'count'},fill_value=0)
pivot_contents = pivot_contents > 0
pivot_contents.head()
```
