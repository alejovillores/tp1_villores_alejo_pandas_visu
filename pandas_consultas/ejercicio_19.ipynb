## Ejercicio 19


Calcule la latitud y longitud promedio de los contenidos con referencias en la tierra y
diga dónde está eso

```python
# Importo las librerias para trabajar con los respectivos datos.
import pandas as pd
import numpy as np
```

```python
geo_tags_base = pd.read_csv("geo_tags.csv")

#Genero una copia del DataFrame por si me equivoco.
geo_tags = geo_tags_base.copy()
geo_tags.head()
```

```python
geo_tags.info()
```

Podemos ver que hay varias columnas con muchos NaN
Pero las que necesito para resolver el ejercicio poseen todos los datos.

```python
# Filtro la columna cuyo gt_globe es de la tierra.
earth_data = geo_tags.loc[geo_tags['gt_globe'] == 'earth']
earth_data.head()
del geo_tags
```

```python
mean_lat = earth_data['gt_lat'].mean()
mean_lon = earth_data['gt_lon'].mean()
print('Latitud promedio de contenidos referidos a Tierra:' ,str(mean_lat))
print('Longitud promedio de contenidos referidos a Tierra:', str(mean_lon))
```
